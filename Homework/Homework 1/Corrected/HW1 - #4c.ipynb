```python
#!/usr/bin/env python

'''Creates a PigLatin generator.'''

__author__ = 'sewinn'
__email__ = "sewinn@ucsc.edu"
__credits__ = []

vowels = 'aeiou'
punct = '.,?!;:'
pig = 'ay'

def findFirstVowel(w):
    """
    Partitions word according to the position
    of the first vowel
    
    Args:
        w: a word
    
    Returns:
        ret: word partitioned according to first vowel
        
        Raises:
            Nothing.
    
    """
    
    if w.isalpha():
        m = len(w) + 1 
        for i in w:    
            if i in vowels:
                c = w.find(i)
                m = min(c, m)
        if m < len(w) + 1:
            ret = m
        else:
            ret = -1
        return ret

def pigLatin(w):
    """
    This funcion handles pigLatin for word w
    
    Args:
        w: a word
        
    Returns:
        pigLatined: a PigLatin representation of word w
        
    Raises:
        Nothing.
        
    """
    #!!! Good.
    if w[-1] in punct:
        p = w[-1:]
        w = w[:-1]
    else:
        p = ''
    if w.isalpha() and len(w) > 1:
        if w[0] in vowels and not 'y':
            ret = w[0:-1]
        elif w == 'if':
            ret = w + 'w'
        else:
            vowelPlace = findFirstVowel(w)
            if vowelPlace == -1:
                ret = w
            else:
                onset = w[:vowelPlace]
                ret = w[vowelPlace:] + onset
        w = ret + pig
        return w + p
    else:
        return w + p

def pigLatinParts(text):
    """
	this function handles pigLatin for all words in string S. It also remembers the punctuation as at the last Goedel number.

    Args:
		text: a string

    Returns:
        pigLatined: what people can download and use (a program)
        
    Raises:
        Nothing

	"""                   
        
    # my suggestion: break this into a few parts:
    # 1) break up the text into tokens
    tokens = text.split()
    text = []
    # 2) run a for loop over each token where you get the pigLatined version of that
    #      for (2), you'll need to call pigLatin on that token
    for i in tokens:
        token = pigLatin(i)
        text.append(token)        
    # 3) put it all together to make one piece of text
    #      for that, you can use another for loop and + (just keep adding to the string you are going to return
    #      OR you can use a function called join []; see the note below
    text = " ".join(text)
    
    return text #not what you want to do here, but this in the business of running.

if __name__ == "__main__":
    
    # all of the below code is used to TEST whether yours worked
    
    inputLine = "if you think, sadly, that he will actually visit, then I am fine."
    key = "ifway ouyay inkthay, adlysay, atthay ehay illway actuallyay isitvay, enthay I amay inefay." 
    pg = pigLatinParts(inputLine)
    if pg ==key:
        result = "PASS"
    else:
        result = "FAIL"
    print inputLine
    print pg
    print result
```
