```python
vowels = 'aeiou'

def findFirstVowel(w):
    """
    Partitions word according to the position
    of the first vowel
    
    Args:
        w: a word
    
    Returns:
        ret: word partitioned according to first vowel
        
        Raises:
            Nothing.
    
    """
    
    if w.isalpha():
        m = len(w) + 1 
        for i in w:    
            if i in vowels:
                c = w.find(i)
                m = min(c, m)
        if m < len(w) + 1:
            ret = m
        else:
            ret = -1
        return ret

def pigLatin(w, d):
    """
    This funcion handles pigLatin for word w, for dialect d
    
    Args:
        w: a word
        d: a dialect (a string)
        
    Returns:
        pigLatined: a PigLatin representation of word w
        
    Raises:
        Nothing.
        
    """
    
    if w.isalpha():
        if w[0] in vowels:
            ret = w[0:-1]
        else:
            vowelPlace = findFirstVowel(w)
            if vowelPlace == -1:
                ret = w
            else:
                onset = w[:vowelPlace]
                ret = w[vowelPlace:] + onset
            if ret[-1] not in vowels:
                """
                if partitioned word ends with a consonant,
                omit the onset of dialect ending
                
                """
                #!!! Excellent. That's my preferred solution to this part too.
                d = d[1:]
                return ret + d
            else:
                return ret + d
        return ret + d
    else:
        return w

if __name__ == "__main__":
    
    # all of the below code is used to TEST whether yours worked
    # a list of test word -> pg examples, as a dict
    wordsToPigLatin = {"artichoke": ("artichokyay", "yay"), 
                       "stipple": ("ipplestay", "way"),
                       "123re": ("123re", "ay"),
                       "soup": ("oupsay", "way")}
    
    # per word, get the pg, the expected answer, and print success
    for word in wordsToPigLatin:
        ans, dialect = wordsToPigLatin[word]
        pg = pigLatin(word, dialect)
        if pg == ans:
            result = "PASS"
        else:
            result = "FAIL"
        print word, pg, result
```
