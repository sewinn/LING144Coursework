```python
#!/usr/bin/env python

'''Creates a PigLatin generator.'''

__author__ = 'sewinn'
__email__ = 'sewinn@ucsc.edu'

vowels = 'aeiou'
pig = 'ay'

def findFirstVowel(w):
    """
    Partitions word according to the position
    of the first vowel
    
    Args:
        w: a word
    
    Returns:
        ret: word partitioned according to first vowel
        
        Raises:
            Nothing.
    
    """
    
    if w.isalpha():
        m = len(w) + 1 
        for i in w:    
            if i in vowels:
                c = w.find(i)
                m = min(c, m)
        if m < len(w) + 1:
            ret = m
        else:
            ret = -1
        return ret

def pigLatin(w):
    """
    This funcion handles pigLatin for word w
    
    Args:
        w: a word
        
    Returns:
        pigLatined: a PigLatin representation of word w
        
    Raises:
        Nothing.
        
    """
    
    if w.isalpha():
        if w[0] in vowels:
            ret = w[0:-1]
        else:
            vowelPlace = findFirstVowel(w)
            if vowelPlace == -1:
                ret = w
            else:
                onset = w[:vowelPlace]
                ret = w[vowelPlace:] + onset
        w = ret + pig
        return w
    else:
        return w

            

if __name__ == "__main__":
    
    wordsToPigLatin = {"artichoke": "artichokay", "stipple": "ipplestay", "123re": "123re", "soup": "oupsay"}
    
    for word in wordsToPigLatin:
        pg = pigLatin(word)
        ans = wordsToPigLatin[word]
        if pg == ans:
            result = "PASS"
        else:
            result = "FAIL"
        print word, pg, result
```
