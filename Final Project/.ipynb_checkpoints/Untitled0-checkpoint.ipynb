```python
from nltk.corpus import wordnet as wn
import re
 
colorDict = {'n':'green',
             'v':'red',
             'a':'blue',
             'r':'yellow'} #Establishes a dictionary which assigns each part of speech a color to be tagged with.
 
def posHighlight(bookPath, partOfSpeech):
   
    #The next four lines open the book file and set up a regular expression search for all words.
    fb = open(bookPath, 'r')
    fb.close
    book = fb.read()
    wordRE = re.compile(r'[\w]+')
    #Note that the search splits contractions into two results ("you're" is returned as "you" and "re"). This is
    #fine, as we'll be using the words looking for words to compare with the parts of speech in their WordNet
    #entries, and contractions do not have clear parts of speech.
 
    #The following two lines give the results of word.RE(book) as strings rather than a single SRE object.
    for matches in wordRE.finditer(book):
        word = book[matches.start():matches.end()]

        #These lines search for an entry of the part of speech given for each word of the given text file, and
        #if it exists, it highlights the particular word in an HTML color tag, and replaces the untagged instance
        #in the given text file.
        if wn.synsets(word, partOfSpeech):
            colorWord = '<color="{0}">{1}</font>'.format(colorDict[partOfSpeech], word)
            output = re.sub(word, colorWord, book)
        return output

posHighlight('.', 'n')
```
